## **Prueba de chi-cuadrado para la selección de características**

Las características adicionales actúan como un ruido para el cual el modelo de Machine Learning puede tener un rendimiento extremadamente bajo.El modelo tarda más tiempo en entrenarse.Asignación de recursos innecesarios para estas características.

```python

from sklearn.datasets import load_iris
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
  
X = df.iloc[:,0:9]  #independent columns
y = df.iloc[:,-1]    #target column i.e price range#apply SelectKBest class to extract top 10 best features
  
# Convert to categorical data by converting data to integers
dfk = df.astype(int)
  
# Two features with highest chi-squared statistics are selected
chi2_prueba = SelectKBest(chi2, k = 9)
entrenamiento= chi2_prueba.fit(X, y)

np.set_printoptions(precision=3) #Tantos despues de la coma
print(entrenamiento.scores_)



# Reduced features
#print('Original feature number:', X.shape[1])
#print('Reduced feature number:', X_kbest_features.shape[1])
```

De acuerdo a las carteristicas con los valores de shi mas altos será descartada la columna 'Possession of Weapons'(5.766).

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
X=df.iloc[:,[1,7]].values
y=df.iloc[:,-1].values
train_X,test_X,train_y,test_y=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
modelo=LinearRegression()
modelo.fit(train_X,train_y)
y_pred=modelo.predict(test_X)

print('Datos de entrenamiento:')
print(test_y)
print('Datos obtenidos en la predicción:')
print(y_pred)
```

```python
from math import sqrt
from sklearn.metrics import mean_squared_error
error=sqrt(mean_squared_error(test_y,y_pred))
print(error)
```

```python
modelo=SVR()
modelo.fit(train_X,train_y)
y_pred=modelo.predict(test_X)

print('Datos de entrenamiento:')
print(test_y)
print('Datos obtenidos en la predicción:')
print(y_pred)
```

```python
from math import sqrt
from sklearn.metrics import mean_squared_error
error=sqrt(mean_squared_error(test_y,y_pred))
print(error)
```
